##  Regresja logistyczna 

Regresja logistyczna jest modelem parametrycznym, a zatem zmienne powinny być w tej samej skali. Przed uczeniem modelu stanadaryzujemy dane ze zbioru testowego. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler_2var = StandardScaler()
X_train_standardized = scaler_2var.fit_transform(X_train)
X_test_standardized = scaler_2var.transform(X_test)
model_lr = LogisticRegression()
model_lr.fit(X_train_standardized, y_train)
```

Stworzymy funkcję obrazującą działanie klasyfikatora.

```python
from numpy import where
from numpy import meshgrid
from numpy import arange
from numpy import hstack

def plot_classification_surface(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # określenie granic zbioru
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1

    # skalowanie dla obu osi
    x1grid = arange(min1, max1, 0.1)
    x2grid = arange(min2, max2, 0.1)

    # utworzenie siatki
    xx, yy = meshgrid(x1grid, x2grid)

    # przetworzenie siatki w wektor
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))

    # stworzenie zmiennych objaśniających dla modelu
    grid = hstack((r1,r2))

    # predykcja modelu zbioru
    yhat = trained_model.predict(grid)

    # przetworzenie predykcji na decyzji dla konkretnego punktu wykresu
    zz = yhat.reshape(xx.shape)

    # wizualizacja siatki z przyporządkowaną klasą
    plt.contourf(xx, yy, zz, cmap='Paired')

    # tworzenie wykresu punktowego dla klas ze zbioru X
    for class_value in range(2):

        # przyporządkowanie klasy (y) do obsewacji (X)
        row_ix = where(y_plot == class_value)

        # stworzenie wykresu punktowego
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], cmap='Paired', alpha=0.3, label=class_value)

    # wizualizacja wykresu
    plt.legend(loc='upper right')
    plt.show()

plot_classification_surface(X_plot=X_train_standardized, y_plot=y_train, trained_model=model_lr)
```

###  Ocena modelu regresji logistycznej

Zbudujemy macierz omyłek dla modelu regresji logistycznej.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

predictions_lr = model_lr.predict(X_test_standardized)
cm = confusion_matrix(y_test, predictions_lr)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

Policzmy dokładność, precyzję, czułość oraz F1 score.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Accuracy: ', accuracy_score(y_test, predictions_lr))
print("Precision:", precision_score(y_test, predictions_lr))
print("Recall:", recall_score(y_test, predictions_lr))
print("F1_score:", f1_score(y_test, predictions_lr))
```

Sporządzimy teraz wykresy krzywej ROC (Receiver Operating Characteristic) i pochodnej tego wykresowi metryki AUC (Area Under the Curve). Modele klasyfikacji mogą zwracać prawdopodobieństwo i to wykorzystuje krzywa ROC. Standardowo modele klasyfikacji mają punkt odcięcia na 0.5, tj. jeśli prawdopodobieństwo jest wyższe niż ten punkt odcięcia to klasa jest pozytywna, w przeciwnym razie negatywna. Można badać zmianę wynikó w zależności od punktu odcięcia. Jeżeli wartość metryki "precision" jest niewystarvzająca, to zamiast uczyć nowy model można spróbować najpierw manipulować punktem odcięcia.

AUC jest to pole znajdujące się pod krzywą ROC, zaś sama krzywa ROC to wykres TPR (True Positives Rate – tzw. czułość) oraz FPR (False Positives Rate – tzw. specyficzność) w zależności od punktu odcięcia. Najlepszym modelem jest ten, który jest najbliżej lewego górnego rogu wykresu, czyli dążymy do TPR równego 1 i FPR równego 0. 

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# prognoza prawdopodobieństwa
predictions_probability_lr = model_lr.predict_proba(X_test_standardized)

# obliczenie FPR oraz TPR w zależności od punktu odcięcia
fpr, tpr, thresholds = roc_curve(y_test, predictions_probability_lr[:,1])

# wizualizacja ROC Curve
fig = plt.figure(figsize=(6, 6))

# rysowanie prostej dla modelu losowego
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()

# obliczenie AUC
auc = roc_auc_score(y_test, predictions_probability_lr[:,1])
print('AUC: ' + str(auc))
```

Stworzymy funkcję, która oblicza najważniejsze metryki oraz rysuje wykresy. 

```python
from sklearn.metrics import classification_report

metrics_dataframe = pd.DataFrame(columns = ['Model', 'F1_score', 'AUC'])
metrics_dataframe
models = []
models_names = []
predictions_proba_list = []
def calculate_metrics(model, name, X_checked, y_checked):
    models.append(model)
    models_names.append(name)
    global metrics_dataframe
    predictions = model.predict(X_checked)
    predictions_proba = model.predict_proba(X_checked)
    predictions_proba_list.append(predictions_proba[:,1])

    ############## metryki dla sprawdzanego modelu ################
    # Precision, Recall, F1, Accuracy
    print(classification_report(y_checked, predictions))

    # Confusion matrix
    plt.figure()
    cm = confusion_matrix(y_checked, predictions)
    ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
    ax.set_title('Confusion Matrix\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    plt.show()

    # plot ROC curve
    fig = plt.figure(figsize=(6, 6))
    plt.plot([0, 1], [0, 1], 'k--')
    for model_selected, name_selected, pred_proba in zip(models, models_names, predictions_proba_list):
        fpr, tpr, thresholds = roc_curve(y_checked, pred_proba)
        plt.plot(fpr, tpr, label=name_selected)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

    f1_metric = f1_score(y_checked, predictions)
    auc_metric = roc_auc_score(y_checked, predictions_proba[:,1])
    metrics_dataframe = metrics_dataframe.append({'Model': name, 'F1_score': f1_metric, 'AUC': auc_metric},
                                                 ignore_index=True)
    return metrics_dataframe
```

```python
calculate_metrics(model_lr, 'Logistic Regression', X_test_standardized, y_test)
```

# KNN (K Nearest Neighbours) – K najbliższych sąsiadów

Potrzebne nam będę ustandaryzowane zmienne, tak aby zmienne miały ten sam zakres wartości. 

Model zapamiętuje dane ze zbioru treningowego i na tym kończy się jego nauka, gdyż nie ma tu żadnych parametrów do wytrenowania, jak ma to miejsce dla regresji logistycznej. Taki model nazywamy leniwym (lazy learner). Obserwacji przyporządkowana jest najliczniejsza klasa pośród k najbliższych obserwacji. Warto zatem, aby k było liczbą nieparzystą, wtedy zawsze będzie wynik korzystny dla jednej z obserwacji.

Sposób mierzenia odległości między obserwacjami jest hiperparametrem modelu KNN: w zależności od problemu inne metryki odległości mogą sprawdzić się lepiej.

Algorytm polega na doborze liczby K i wyborze metryki odległości. Następnie dla obserwacji, dla której wykonujemy prognozę, wyliczamy odległości od wszystkich obserwacji ze zbioru treningowego. Znajdujemy k najbliższych obserwacji, a następnie przypisujemy większościową klasę. Od parametru K zależy czy model będzie niedouczony czy przeuczony.

Liczba K jest hiperparametrem. Wybierzmy K jako 0.1% wszystkich obserwacji (będzie to K = 10).

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors= int(X_train_standardized.shape[0]*0.001))
model_knn.fit(X_train_standardized, y_train)
plot_classification_surface(X_plot=X_train_standardized, y_plot=y_train, trained_model=model_knn)
```

Możesz zauważyć, że wytrenowany model KNN nie jest separatorem liniowym, jak i zauważalne jest przeuczenie modelu.

Oceńmy wytrenowany KNN, korzystając z stworzonej przez nas funkcji calculate_metrics.

```python
calculate_metrics(model_knn, 'KNeighborsClassifier', X_test_standardized, y_test)
```

# Boosting — AdaBoost

Nazwa pochodzi od angielskiego stwierdzenia Adaptive Boosting. Boosting jest metodąpolegającą na składaniu kilku klasyfikatorów razem w celu uzyskania lepszego wyniku. Istnieje jedna różnica, która powoduje, iż AdaBoost działa zazwyczaj lepiej niż RandomForest.

    RandomForest szkoli wiele klasyfikatorów jednocześnie/równolegle, każdy na podzbiorze danych wejściowych, następnie uśrednia wynik.
    AdaBoost szkoli wiele klasyfikatorów sekwencyjnie/po sobie, przekazując dane odpowiednio zmodyfikowane/przerobione przez poprzedni klasyfikator, do kolejnego. Następny klasyfikator skupia się na danych źle zaklasyfikowanych przez poprzednika.

Klasyfikatorowi AdaBoost przekazujemy:

    base_estimator - klasyfikator, którego ma użyć AdaBoost (dowolny z sklearn -> klasyfikacji) np. DecisionTree,
    n_estimators - liczba klasyfikatorów (ekspertów), które mają być połączone sekwencyjnie,
    pozostałe parametry służą do poprawy wyniku i wymagają ustawienia.



```python
from sklearn.ensemble import AdaBoostClassifier

model_adaboost = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50)
model_adaboost.fit(X_train, y_train)
plot_classification_surface(X_plot=np.array(X_train), y_plot=y_train, trained_model=model_adaboost)
```

```python
calculate_metrics(model_adaboost, 'Adaboost', X_test, y_test)
```

# Ensembling – Zespół klasyfikatorów

Ensembling, czyli składanie, jest bardzo popularną techniką polegającą na złożeniu wyników kilku modelu.

Można to zrobić przykładowo w następujący sposób:

    Voting Ensemble – uśredniamy wyniki wszystkich użytych przez nas klasyfikatorów. Wszystkie klasyfikatory posiadają takie same wagi np. dla pięciu klasyfikatorów każdy będzie miał wagę 0.2.
    Różne wagi dla klasyfikatorów – tutaj zaczyna się zabawa. W tym przypadku ustalamy wagi dla każdego z klasyfikatorów osobno, indywidualnie, tak aby ich suma wynosiła 1.0.



```python
from sklearn.ensemble import VotingClassifier

model_voting = VotingClassifier(estimators=[('Tree', model_tree),
                                            ('Random Forest', model_forest),
                                            ('AdaBoost', model_adaboost)],
                                voting='soft')

model_voting.fit(X_train, y_train)
plot_classification_surface(X_plot=np.array(X_train), y_plot=y_train, trained_model=model_voting)
```

```python
calculate_metrics(model_voting, 'VotingClassifier', X_test, y_test)
```
