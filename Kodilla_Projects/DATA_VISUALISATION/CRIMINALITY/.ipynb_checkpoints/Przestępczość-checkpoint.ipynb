# Analysis of Fatal Police Shootings in the USA

## Downloading data and their initial analysis

We begin with downloading the relevant breakdown of police fatalities in the US. We will look at some sample data collected in the table.

```python
# Download database of US police fatalities and then load it into a DataFrame.
import pandas as pd
df = pd.read_csv('fatal_police_shootings_data.csv')
df.head()
```

The table above contains information about the victim's name and surname, date and manner of death, and information about the type of weaponry. Additional personal information collected in the table is: gender, race, age and city and state of the event. Details about the incident were also provided: whether the victim showed signs of mental illness, the risk of attack, and whether the victim tried to escape. There is also information whether the event was recorded.

## Race and signs of mental illness of victims

Using pivot tables, we will provide a summary of the relationship between race and whether the victim showed signs of mental illness.

```python
# Reshape the table so that it shows the number of victims of the intervention by 'race'
# and whether they showed ‘signs_of_mental_illness’
import numpy as np

df.pivot_table(values = 'id', index = 'signs_of_mental_illness', columns = 'race',aggfunc='count')
```

Abbreviations
+ A - Asian
+ B - Black
+ H - Hispanic or Latino
+ N - Native Hawaiian/Pacific Islander
+ O - Other
+ W - White

In the next two tables, we will give a summary of the percentage of victims indicating signs of mental illness during the intervention. We will use the built-in crosstab function of the pandas library. We will present a vertical and horizontal version.



```python
pd.crosstab(df['race'], df['signs_of_mental_illness'], values=df['id'], aggfunc=np.sum, normalize='index')
```

```python
pd.crosstab(df['signs_of_mental_illness'], df['race'], values=df['id'], aggfunc=np.sum, normalize='columns')
# Wersja horyzonatalna
```

Conclusion: the highest percentage of victims showing symptoms of mental illness is among Caucasians (26.4%).

## Liczba ofiar a dzień tygodnia 

W tej części przedstawimy wykres słupkowy liczby incydentów w zależności od dnia tygodnia.
Wykorzystamy bibliotekę calendar oraz matplotlit.pyplot

```python
# Dodaj kolumnę oznaczającą dzień tygodnia, w którym doszło do interwencji. 
# Zlicz interwencje według odpowiedniego dnia tygodnia. 
# Następnie stwórz wykres kolumnowy, 
# tak aby dni tygodnia były uszeregowane od poniedziałku do niedzieli.

from calendar import day_name
import matplotlib.pyplot as plt
# dodawanie kolumny 
df['date'] = pd.to_datetime(df['date'])
df['day_name'] = df['date'].dt.day_name()
# wprowadzenie nazw dni tygodnia
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = pd.Categorical(df['day_name'], categories=cats, ordered=True)
df = df.sort_values('day_name')
# zliczenie dni tygodnia
df.groupby('day_name')['id'].count()
# wykres słupkowy
plt.bar(cats, df.groupby('day_name')['id'].count(),
        width = 0.4)

```

## Liczba śmiertelnych interwencji na 1000 mieszkańcówwzględem poszczególnych stanów 

W ostatniej części zajmiemy się analizą ilości incydentów na 1000 mieszkańców w poszczególnych stanach. 
Miara ta jest zdecydowanie lepsza niż wartość bezwzględna ilości incydentów, 
gdyż uwzględnia liczbę mieszkańców danego stanu.

Dane dotyczące stanu zdarzenia są podane w formie skrótu, 
stąd też konieczność połączenia dostępnej bazy danych z bazą danych zawierającą objaśnienie tego skrótu.

Kolejne kroki zostaną wyjaśnione w komentarzach zawartych w kodzie.

```python
# Wczytaj do projektu dane dotyczące populacji w poszczególnych stanach USA oraz 
# dane dotyczące skrótów poszczególnych stanów. 
# Połącz te bazy danych w taki sposób, aby móc policzyć do ilu incydentów w bazie 
# dotyczącej śmiertelnych interwencji doszło w przeliczeniu na 1000 mieszkańców każdego ze stanów.
# -------------------------------------------------------
# Zliczenie liczby incydentów i zmiana  nazwy kolumny
df_by_state = df.groupby('state')['id'].count()
df_by_state.rename_axis("Abbreviation" ,inplace=True)
```

```python
# Wczytanie populacji
pop_states_all = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population', header=0)
col_name = pop_states_all[0].columns[3]
pop_states_all[0].rename(columns={col_name: 'Population'},inplace=True)
pop_states = pop_states_all[0][['State', 'Population']].sort_values(by='State') 
#pop_states_all[0].iloc[:,3]  - działamy na indeksie 
```

```python
# Wczytanie skrótów
abv_states_all = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations', header=0)
col_name_state = abv_states_all[0].columns[0]
col_name_abv = abv_states_all[0].columns[5]
# Połączenie skrótów i liczebności populacji
abv_states_all[0].rename(columns={col_name_state: 'State', col_name_abv: 'Abbreviation'},inplace=True) 
abv_states = abv_states_all[0][['State', 'Abbreviation']].sort_values(by='State')
df_pop_partial = pd.merge(pop_states, abv_states,how='inner',on='State' ) 
# Połączenie skrótów i liczebności populacji z ilością incydentów
df_pop = pd.merge(df_pop_partial, df_by_state ,how='inner',on='Abbreviation' )
# Dodanie kolumny: ilość incydentów na 1000 mieszkańców
df_pop['Per 1000'] = (df_pop.id / df_pop.Population ) *100
df_pop
```

```python
df_pop['Per 1000'].max()
```

Wniosek. Najwyższa liczba incydentów na 1000 mieszkańców jest na Alasce i wynosi 0.0053.
